```python
# 5. Given is a dataframe showing name, occupation,salary of people. Find the average salary per occupation.

import pandas as pd
emp={'name':['Adarsh','Vishnu','Dathan','Rohit'],
      "Occupation":['Manager','Programmer','Analyst','Analyst'],
      "Salary" : [23000,45000,33000,35000]}
df = pd.DataFrame(emp)
print(df)
avg_s = df.groupby('Occupation')['Salary'].mean()
print("\nThe average salary for each occupation is: ")
print(avg_s)
```
